```python
import pandas as pd
from ast import literal_eval
```

```python
df = pd.read_csv('../files/prprocessed.csv', encoding='latin1')
```

```python
df.shape
```

```python
df = df[df['crew'].apply(lambda x: "'Director', 'name': " in x)]
```

```python
df.shape
```

```python
cast = df.cast.apply(literal_eval)
```

```python
def get_actors(l):
    acts = ''
    c = 0
    for d in l:
        c += 1
        acts += d['name'] + '|'
        
        if  c >= 3:
            break
        
    return acts[:-1]
```

```python
get_actors(cast[0])
```

```python
actors = cast.apply(get_actors)
```

```python
directors = df['crew'].apply(lambda x: x.split("'Director', 'name': ")[1].split(',')[0][1:-1])
```

```python
posterId = df.imdb_id.apply(lambda x : str(x) + '.jpg')
```

```python
df['director'] = directors
df['actors'] = actors
df['posterID'] = posterId
```

```python
new_df = df[['posterID', 'Genre', 'overview', 'IMDB Score', 'director', 'actors', 'runtime', 'num_faces', 'brightness',
       'saturation', 'hue', 'brightness_sd', 'saturation_sd ', 'hue_sd',
       'blue', 'blue_sd', 'green', 'green_sd', 'red', 'red_sd ']]
```
